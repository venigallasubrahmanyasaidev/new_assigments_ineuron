1) Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
A=int(input("Enter value to check positive,Negative or Zero:"))

if A==0:
    print("the number is Zero")
elif A>0:
    print("the number is postive number")
else:
    print("the number is Negative number")
            
```

2) Write a Python Program to Check if a Number is Odd or Even?

```python
A=int(input("Enter value to check Odd or even:"))

if A%2==0:
    print("number is even")
else:
    print("number is odd")
```

3) Write a Python Program to Check Leap Year?

```python
year=int(input("enter a year to check Leap year or not: "))
if year%4==0 and year%100==0 and year%400==0:
    print("leap year:",year)
else:
    print("Not a leap year:",year)
```

4)Write a Python Program to Check Prime Number?

```python
num=int(input("enter the number to check prime or not: "))

if num>1:
    for i in range(2,(int(num/2))+1):
        if num%i==0:
            print("not a prime number:",num)
            break
    else:
        print("prime number:",num)

    
```
